```python
import os.path as path
import numpy as np
from scipy.misc import imread
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
imgpath = '/data/comp-vision/test6.jpg'
img = imread(imgpath)
plt.figure(figsize=(10, 6))
plt.imshow(img)
```

## Grayscale

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.figure(figsize=(10, 6))
plt.imshow(gray, cmap='gray')
```

bin_gray = np.zeros_like(gray)
bin_gray[(gray >= 180) & (bin_gray <= 255)] = 1
plt.figure(figsize=(10, 6))
plt.imshow(bin_gray, cmap='gray')

## RGB Channels

```python
reds = img[:, :, 0]
blues = img[:, :, 1]
greens = img[:, :, 2]
plt.figure(figsize=(20, 20))

fig1 = plt.subplot(131)
fig1.imshow(reds, cmap='gray')
fig1.set_title('reds')

fig2 = plt.subplot(132)
fig2.imshow(blues, cmap='gray')
fig2.set_title('blues')

fig3 = plt.subplot(133)
fig3.imshow(greens, cmap='gray')
fig3.set_title('greens')
```

```python
bin_red = np.zeros_like(reds)
bin_red[(reds >= 200) & (reds <= 255)] = 1
plt.figure(figsize=(10, 6))
plt.imshow(bin_red, cmap='gray')
```

## HLS Channels

```python
hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
hues = hls[:, :, 0]
lights = hls[:, :, 1]
sats = hls[:, :, 2]

plt.figure(figsize=(20, 20))

fig1 = plt.subplot(131)
fig1.imshow(hues, cmap='gray')
fig1.set_title('hues')

fig2 = plt.subplot(132)
fig2.imshow(lights, cmap='gray')
fig2.set_title('lights')

fig3 = plt.subplot(133)
fig3.imshow(sats, cmap='gray')
fig3.set_title('sats')
```

```python
bin_sats = np.zeros_like(sats)
bin_sats[(sats >= 90) & (sats <= 255)] = 1
plt.figure(figsize=(10, 6))
plt.imshow(bin_sats, cmap='gray')
```
