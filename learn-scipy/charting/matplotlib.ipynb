```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Even though this object is not used explicitly, it is needed for 3d plots
%matplotlib inline
```

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cosx, sinx = np.cos(x), np.sin(x)
```

# Single Plot

```python
# Basic graph - note calling plt.plot over and over again causes the lines to be part of the same plot.
# plt.plot docs - http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot
# plt.figure(figsize=(10, 6))  --> Now that I am using notebook mode I can resize the plot interactively, no need to set this
# plt.plot(x, cosx, color='blue', linewidth=1.0, linestyle='-', label='cos(x)')
plt.plot(x, cosx, color='blue', linewidth=1.0, linestyle='-', label=None)
plt.plot(x, sinx, color='green', linewidth=1.0, linestyle='-', label='sin(x)')
plt.legend(loc='upper left')
plt.title('Sine and Cosine Waves')
```

```python
# Basic graph
# plt.figure(figsize=(10, 6)) --> no need for this anymore
plt.figure()  # If I don't initialize a new figure, it will keep updating the last plot
plt.plot(x, cosx, color='red', linewidth=1.0, linestyle='-', label='cos(x)')
plt.plot(x, sinx, color='yellow', linewidth=1.0, linestyle='-', label='sin(x)')
plt.legend(loc='upper left')

# Change limits
plt.xlim(-5, 5)
plt.ylim(-1.5, 1.5)

# Add a grid
plt.grid(True)
```

```python
# Basic graph
# plt.figure(figsize=(10, 6))
plt.figure()
plt.plot(x, cosx, color='blue', linewidth=1.0, linestyle='-', label='cos(x)')
plt.plot(x, sinx, color='green', linewidth=1.0, linestyle='-', label='sin(x)')
plt.legend(loc='upper left')

# Change limits
plt.xlim(-5, 5)
plt.ylim(-1.5, 1.5)

# Add a grid
plt.grid(True)

# Change the axes to 0,0
ax = plt.gca()  # gca => get current axis
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.spines['bottom'].set_position('zero')
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
```

# Multiple Plots

```python
# plt.figure(figsize=(10, 6))
plt.figure()

fig1 = plt.subplot(211) # ==> plot with 2 rows, 1 col, 1st plot
fig1.plot(x, cosx, color='blue', linewidth=1.0, linestyle='-', label='cos(x)')
fig1.legend(loc='upper left')
fig1.set_title('sine curve')

fig2 = plt.subplot(212) # ==> plot with 2 rows, 1 col, 2nd plot
fig2.plot(x, sinx, color='green', linewidth=1.0, linestyle='-', label='sin(x)')
fig2.legend(loc='upper left')
```

```python
# plt.figure(figsize=(10, 6))
plt.figure()

fig1 = plt.subplot(211) # ==> plot with 2 rows, 1 col, 1st plot
fig1.plot(x, cosx, color='blue', linewidth=1.0, linestyle='-', label='cos(x)')
fig1.legend(loc='upper left')

# Change the xlim and ylim of the 1st plot. Note, now I have to call set_ylim instead of just ylim.
fig1.set_ylim(-1.5, 1.5)
fig1.set_xlim(-3.5, 3.5)

# Add a grid to the first plot
fig1.grid(True)

fig2 = plt.subplot(212) # ==> plot with 2 rows, 1 col, 2nd plot
fig2.plot(x, sinx, color='green', linewidth=1.0, linestyle='-', label='sin(x)')
fig2.legend(loc='upper left')
```

# 3D Plots

```python
# Viewing this in qt-mode will let me turn the plot
x = np.random.randint(-10, 10, 100)
y = np.random.randint(-10, 10, 100)
z = np.random.randint(-10, 10, 100)

plt.figure()
fig = plt.subplot(111, projection='3d')
fig.scatter(x, y, z)
fig.set_xlabel('X Label')
fig.set_ylabel('Y Label')
fig.set_zlabel('Z Label')
```

# Format Axis

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cosx, sinx = np.cos(x), np.sin(x)

fig1 = plt.subplot(121)
# fig1.plot(x, cosx)
fig1.plot(x, cosx, color='blue', linewidth=1.0, linestyle='-', label='cos(x)')
fig1.legend(loc='upper left')
# fig1.set_title('Cosine', fontdict={'color':[0, 1, 0], 'size': 16})
fig1.set_title('Cosine', fontdict={})
fig1.tick_params(color='w', labelcolor='w')
# fig1.set_xlabel('x', fontdict={'color': 'r', 'size': 14})
fig1.set_xlabel('x')
fig1.set_ylabel('cos(x)', fontdict={'color': 'w', 'size': 14})

fig2 = plt.subplot(122)
fig2.plot(x, sinx)
fig2.set_title('Sine', fontdict={'color':'w', 'size': 16})
fig2.tick_params(color='w', labelcolor='w')
fig2.set_xlabel('x', fontdict={'color': 'r', 'size': 14})
fig2.set_ylabel('sin(x)', fontdict={'color': 'w', 'size': 14})
```
