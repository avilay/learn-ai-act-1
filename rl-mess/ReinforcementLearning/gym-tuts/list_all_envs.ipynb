```python
allentrypoints = set()
for envspec in allenvs:
    ep = envspec.entry_point.split(":")[0]
    lib = ".".join(ep.split(".")[:3])
    allentrypoints.add(lib)
```

```python
allentrypoints
```

```python
good_entry_points = {
    'gym.envs.algorithmic',
    'gym.envs.atari',
    'gym.envs.box2d',
    'gym.envs.classic_control',
    'gym.envs.toy_text',
}
blacklist = ["Defender"]

good_envspecs = []
for envspec in allenvs:
    skip = False
    for bad in blacklist:
        if envspec.id.startswith(bad):
            skip = True
            break
    if skip: continue
    ep = envspec.entry_point.split(":")[0]
    lib = ".".join(ep.split(".")[:3])
    if lib in good_entry_points:
        good_envspecs.append(envspec)
```

```python
len(good_envspecs)
```

```python
good_envspecs[100:110]
```

```python
with open("env-details.txt", "wt") as f:
    for i, envspec in enumerate(good_envspecs):
        print(f"{datetime.now()} {i} examining {envspec.id}")
        env = envspec.make()
        print(f"{envspec.id} ActionSpace={env.action_space} StateSpace={env.observation_space}", file=f)
```
